```python
text = """
On 12/08/2025, the user <user_id=1024> logged into the system at 10:45 AM
and posted: "I scored 98.5% in AI & Data-Science!!! ".
The system recorded CPU usage = 76%, RAM = 8GB, and temperature = 42.3°C.
Later, an alert <warning level="high">Memory-Leak detected</warning> was triggered.
Visit https://example.com/login or email support@example.com for help.
"""
```

### Numbers Conversion to String

```python
pip install num2words
```

```python
from num2words import num2words

def convert_numbers_to_words(text):
    return re.sub(r'\d+', lambda x: num2words(x.group()), text)

text = convert_numbers_to_words(text)

print(text)
```

### Tokenization and Cleaning

```python
sentences = sent_tokenize(text)
print ("Sentence tokens :")
print (sentences)
```

```python
words = word_tokenize(text)
print ("word tokens :")
print (words)
```

```python
text_lower = text.lower()
print("\nLowercase Text:")
print(text_lower)
```

```python
text_clean = re.sub(r'http\s+','',text_lower)# to remove URLs
taxt_clean = re.sub(r'\s+@\s+','',text_clean)# to remove emailIDs
text_clean = re.sub(r'<.*?>','',text_clean)# to remove HTML tags
text_clean = re.sub(r'[^a-z0-9\s]','',text_clean)# to remove special characters
text_clean = re.sub(r'\s+', ' ', text_clean).strip()# to remove extra spaces

print("\nCleaned Text:")
print(text_clean)
```

```python
clean_tokens = word_tokenize(text_clean)
print("\nClean Tokens:")
print(clean_tokens)
```

```python
stop_words = set(stopwords.words('english')) 
#This loads a predefined list of English stopwords from NLTK and Converts the stopword list into a set

filtered_words = [word for word in clean_tokens if word not in stop_words]

print("\nAfter Stopword Removal:")
print(filtered_words)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_words]

print("\nStemmed Words:")
print(stemmed_words,"\n")
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word,pos='v') for word in filtered_words]
#pos is Part Of Speech

print("\nLemmatized Words:")
print(lemmatized_words)
```
